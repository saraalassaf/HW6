```r
data_frame <- data.frame(
    Id = c(1,2,3,4,5),
    name = c("Peter","Amy","Ryan","Gary","Michelle"),
    salary = c(623.30,515.20,611.00,729.00,843.25)
)
data_frame
new_column <- cbind(data_frame, department = c("IT","finance","computer science","marketing","accounting"))
new_column
data_frame[c(1,3,5),c(2,3)]
barplot(data_frame[c(1,4,5),3],names.arg = c("Peter","Gary","Michelle"))
labels <- c("Peter","Amy","Michelle")
pie(data_frame[c(1,2,5),3], label = labels, main = "Our Salaries")
```
